```python
# importing yolo model
from ultralytics import YOLO
```

```python
# initializing the model
model = YOLO('yolov8m.pt')
```

```python
DATA_CFG = '/content/drive/MyDrive/OCR/data/dataset.yaml'
```

```python
# training the model
results = model.train(
    data = DATA_CFG,
    batch = 24,
    epochs = 35,
    device = 0,
    lr0 = 3e-4,
    patience = 0,
    optimizer = 'AdamW',
    verbose = True,
    imgsz = 640,
    save_period = 10
    )
```

## Save Model

```python
import os
from shutil import copy2
```

## Save Logs

```python
dst_dir = "/content/drive/MyDrive/OCR/logs"
src_dir = "/content/runs/detect/train"

log_files = [file_name for file_name in os.listdir(src_dir) if file_name != "weights"]
```

```python
for file_name in log_files:
    src_path = os.path.join(src_dir, file_name)
    dst_path = os.path.join(dst_dir, file_name)

    copy2(src_path, dst_path)
```
